### 【問題1】べき乗の算術演算子を使用した実装

```python
THICKNESS = 0.00008

folded_thickness = THICKNESS * 2**43

print("厚さ： {}メートル".format(folded_thickness))
```

### 【問題2】単位の変換

```python
print("厚さ： {:.2f}万キロメートル".format(folded_thickness / 1000 / 10000))
```

### 【問題3】for文を使用して作成

```python
folded_thickness = THICKNESS

for _ in range(43):
    folded_thickness *= 2

print("厚さ： {}メートル".format(folded_thickness))
```

### 【問題4】計算時間の比較

```python
import time
```

```python
# 「べき乗の算術演算子を使用した実装」の実行時間
start = time.time()

folded_thickness = THICKNESS * 2**43

elapsed_time = time.time() - start
print("time : {}[ms]".format(elapsed_time * 1000))
```

```python
# 「for文を使用した実装」の実行時間
start = time.time()

folded_thickness = THICKNESS

for _ in range(43):
    folded_thickness *= 2

elapsed_time = time.time() - start
print("time : {}[ms]".format(elapsed_time * 1000))
```

### 【問題5】リストへの保存

```python
folded_thickness_list = [THICKNESS]

for _ in range(43):
    folded_thickness_list.append(folded_thickness_list[-1] * 2)

print("リストの要素数： {}個".format(len(folded_thickness_list)))
```

### 【問題6】折れ線グラフの表示

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list)
plt.show()
```

35回を過ぎたあたりから紙の厚さが急激に増加している。

### 【問題7】グラフのカスタマイズ

```python
# グラフにマーカーを追加する。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, markersize=4, marker='o')
plt.show()
```

```python
# グラフにグリッドを追加する
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list)
plt.grid()
plt.show()
```

```python
# y軸を対数軸にする。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list)
plt.yscale('log')
plt.show()
```
